# DL, Exercise 1: Naive Convolution

The goal here is to take an image $I$ and convolve it with a kernel $K$.

We will implement the operation "naively," i.e., using four nested loops.

First, let's read an image.

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

im  = Image.open("dog.jpg")
np_img = np.asarray(im)

plt.imshow(np_img)
plt.show()
```

```python
# Convert to grayscale by max of R, G, B

gray_img = np_img.max(axis=2)

plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
sobel_x = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])

def naive_conv(I, K):
    rows = I.shape[0]
    cols = I.shape[1]
    k_rows = K.shape[0]
    k_cols = K.shape[1]
    S = np.zeros(((rows - k_rows + 1), (cols - k_cols + 1)))
    for i in range(S.shape[0]):
        for j in range(S.shape[1]):
            sum = 0
            for m in range(k_rows):
                for n in range(k_cols):
                    # Range of m, n is 0-2
                    sum += I[i-m+2, j-n+2] * K[m,n]
            S[i,j] = sum
    return S

S = naive_conv(gray_img, sobel_x)
```

```python
plt.imshow(S, cmap='gray')
plt.show()
```
